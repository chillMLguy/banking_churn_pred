## Exploratory Data Analysis (EDA)

## 1. Introduction

In this notebook, we perform Exploratory Data Analysis (EDA) on the bank customer churn dataset https://www.kaggle.com/datasets/radheshyamkollipara/bank-customer-churn.

The goal is to:

- Understand the structure of the data

- Check data quality (missing values, distributions)

- Explore relationships between variables

- Identify patterns related to customer churn (Exited)

This step is very important before building any machine learning models.

## 2. Importing Libraries and Loading the Dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn-v0_8')

data = pd.read_csv("data.csv")
```

## 4. Target Variable Analysis (Exited)

### 4.1 Class Distribution

```python
sns.countplot(x='Exited', data=data)
plt.title('Churn Distribution')
plt.show()
```

## 5. Univariate Analysis (Distributions)

### 5.1 Numerical Feature Distributions

```python
num_cols = data.select_dtypes(include=['int64', 'float64']).columns

for col in num_cols:
    plt.figure(figsize=(6,4))
    sns.histplot(data[col], kde=True)
    plt.title(f'Distribution of {col}')
    plt.show()
```

## 6. Feature vs Target Analysis

### 6.1 Numerical Features vs Exited

Here we show how numerical variables differ between churned and non-churned customers.

```python
for col in num_cols:
    if col != 'Exited':
        plt.figure(figsize=(6,4))
        sns.boxplot(x='Exited', y=col, data=data)
        plt.title(f'{col} vs Exited')
        plt.show()
```

### 6.2 Categorical Features vs Exited

```python
cat_cols = ['Geography', 'Gender', 'Card Type', 'Complain']

for col in cat_cols:
    plt.figure(figsize=(6,4))
    sns.countplot(x=col, hue='Exited', data=data)
    plt.title(f'{col} vs Exited')
    plt.xticks(rotation=30)
    plt.show()
```

## 7. Correlation Analysis

### 7.1 Correlation Matrix

```python
plt.figure(figsize=(12,8))
corr = data[num_cols].corr()
sns.heatmap(corr, annot=False, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

### 7.2 Most Correlated Features with Exited

```python
corr_target = corr['Exited'].sort_values(ascending=False)
corr_target

corr_target.drop('Exited').head(10).plot(kind='bar')
plt.title('Top Correlated Features with Exited')
plt.show()
```
